```python
# Crear un objeto bytearray vacío
bytearray()  # Objeto bytearray vacío
bytearray(b'')  # Objeto bytearray vacío también

# Crear una instancia de bytearray rellena de ceros con la longitud especificada
bytearray(10)  # Instancia rellena de ceros con longitud 10
bytearray(b'\x00\x00\x00\x00\x00\x00\x00\x00\x00\x00')  # Los ceros están representados en bytes

# Crear un bytearray a partir de una secuencia de enteros
bytearray(range(5))  # bytearray a partir de una secuencia de enteros del 0 al 4
bytearray(b'\x00\x01\x02\x03\x04')  # Representación en bytes de los enteros

# Crear un bytearray a partir de una secuencia de bytes
name = bytearray(b'Lina')  # Un bytearray a partir de la secuencia de bytes 'Lina'

# Reemplazar 'L' por 'l' en el bytearray 'name'
name.replace(b'L', b'l')  # Reemplaza 'L' por 'l' en 'name'
bytearray(b'lina')  # El resultado es 'lina'

# Comprobar si el bytearray 'name' termina con 'na'
name.endswith(b'na')  # True, 'name' termina con 'na'

# Convertir 'name' a mayúsculas
name.upper()  # 'name' en mayúsculas es 'LINA'
bytearray(b'LINA')  # El resultado es 'LINA'

# Contar cuántas veces aparece 'L' en el bytearray 'name'
name.count(b'L')  # 1, 'L' aparece una vez en 'name'
```

```python
# Crear un diccionario vacío
d = {}

# Agregar un par (clave, valor) al diccionario
d['a'] = 1  # Establecemos un par (clave, valor) con 'a' como clave y 1 como valor
d['b'] = 2  # Establecemos otro par con 'b' como clave y 2 como valor

# ¿Cuántos pares tiene el diccionario?
len(d)  # 2, porque hay dos pares (clave, valor) en el diccionario

# ¿Cuál es el valor asociado a la clave 'a'?
d['a']  # 1, porque 'a' tiene un valor de 1 en el diccionario

# ¿Cómo se ve el diccionario `d` ahora?
d  # {'a': 1, 'b': 2}, muestra el contenido actual del diccionario

# Eliminar la clave 'a' del diccionario
del d['a']  # Eliminamos la clave 'a' y su valor del diccionario

# ¿Cómo se ve el diccionario `d` después de eliminar 'a'?
d  # {'b': 2}, ahora solo queda la clave 'b' con su valor

# Agregar un nuevo par (clave, valor) al diccionario
d['c'] = 3  # Agregamos 'c' como clave y 3 como valor al diccionario

# ¿La clave 'c' está presente en el diccionario?
'c' in d  # True, 'c' es una clave en el diccionario

# ¿El valor 3 está presente en el diccionario?
3 in d  # False, 3 no es una clave, se verifica si es una clave

# ¿La clave 'e' está presente en el diccionario?
'e' in d  # False, 'e' no es una clave en el diccionario

# Limpiar completamente el diccionario, eliminando todos los pares (clave, valor)
d.clear()  # Eliminamos todos los elementos del diccionario

# ¿Cómo se ve el diccionario `d` después de borrar todo?
d  # {}, el diccionario está vacío
```

```python
# scopes3.py
# Local, Enclosing and Global

def enclosing_func():
    M = 13

    def local():
        # m doesn't belong to the scope defined by the local
        # function so Python will keep looking into the next
        # enclosing scope. This time M is found in the enclosing
        # scope
        print(M, 'printing from the local scope')

    # calling the function local
    local()
    m = 5
    print(M, 'printing from the global scope')

enclosing_func()
```

```python
# bike.py

# Definimos la clase Bike (Bicicleta)
class Bike:
    def __init__(self, colour, frame_material):
        self.colour = colour  # Establecemos el color de la bicicleta
        self.frame_material = frame_material  # Establecemos el material del marco

    def brake(self):
        print("Braking!")  # Imprimimos "Frenando!"

# Creamos un par de instancias de la clase Bike (bicicleta)
red_bike = Bike('Red', 'Carbon fiber')  # Bicicleta roja de fibra de carbono
blue_bike = Bike('Blue', 'Steel')  # Bicicleta azul de acero

# Inspeccionamos los objetos que tenemos, que son instancias de la clase Bike (bicicleta).
print(red_bike.colour)  # Imprime: Red (rojo)
print(red_bike.frame_material)  # Imprime: Carbon fiber (fibra de carbono)
print(blue_bike.colour)  # Imprime: Blue (azul)
print(blue_bike.frame_material)  # Imprime: Steel (acero)

# Frenamos una de las bicicletas
red_bike.brake()  # Imprime: Braking! (¡Frenando!)

```

```python
from collections import ChainMap
default_connection = {'host': 'localhost', 'port': 4567}
connection = {'port': 5678}
conn = ChainMap(connection, default_connection) # map creation
conn['port'] # port is found in the first dictionary
5678
conn['host'] # host is fetched from the second dictionary
'localhost'
conn.maps # we can see the mapping objects
[{'port': 5678}, {'host': 'localhost', 'port': 4567}]
conn['host'] = 'packtpub.com' # let's add host
conn.maps
[{'port': 5678, 'host': 'packtpub.com'},
{'host': 'localhost', 'port': 4567}]
del conn['port'] # let's remove the port information
conn.maps
[{'host': 'packtpub.com'}, {'host': 'localhost', 'port': 4567}]
conn['port'] # now port is fetched from the second dictionary
4567
dict(conn) # easy to merge and convert to regular dictionary
{'host': 'packtpub.com', 'port': 4567}
```

```python
def matrix_mul(a, b):
    return [[sum(i * j for i, j in zip(r, c)) for c in zip(*b)] for r in a]

# Ejemplo de matrices a y b
a = [[1, 2], [3, 4]]
b = [[5, 1], [2, 1]]

# Multiplicación de matrices usando la función matrix_mul
c = matrix_mul(a, b)

# Imprimir el resultado
for row in c:
    print(row)
```

```python
# scoping.level.2.py
def outer():
    test = 1  # variable en el ámbito de outer
    def inner():
        nonlocal test  # declarar test como no local para modificar la variable en outer
        test = 2  # modificar la variable en el ámbito de inner
        print('inner:', test)
    inner()
    print('outer:', test)

test = 0  # variable en el ámbito global
outer()
print('global:', test)
```

```python
# scoping.level.1.py
def my_function():
    test = 1 # this is defined in the local scope of the function
    print('my_function:', test)

test = 0 # this is defined in the global scope
my_function()
print('global:', test)
```
